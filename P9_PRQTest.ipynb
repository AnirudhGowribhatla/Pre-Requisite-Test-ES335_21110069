```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt

x = [0.5]
p = 0.5
while (p!=100.0):
    p = p + 0.5
    x.append(p)
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
def f1(x):
    return x
def f2(x):
    return mt.pow(x,2)
def f3(x):
    return ((x**3)/100)
def f4(x):
    return mt.sin(x)
def f5(x):
    return mt.sin(x)/x
def f6(x):
    return mt.log(x)
def f7(x):
    return mt.exp(x)

for i in x :
    y1.append(f1(i))
    y2.append(f2(i))
    y3.append(f3(i))
    y4.append(f4(i))
    y5.append(f5(i))
    y6.append(f6(i))
    y7.append(f7(i))

plt.plot(x,y1,color = 'b',label = 'y = x')
plt.legend()
plt.show()
plt.plot(x,y2,color = 'r',label = 'y = x^2')
plt.legend()
plt.show()
plt.plot(x,y3,color = 'g',label = 'y = x^3/100')
plt.legend()
plt.show()
plt.plot(x,y4,color = 'c',label = 'y = sin(x)')
plt.legend()
plt.show()
plt.plot(x,y5,color = 'm',label = 'y = sin(x)/x')
plt.legend()
plt.show()
plt.plot(x,y6,color = 'y',label = 'y = log(x)')
plt.legend()
plt.show()
plt.plot(x,y7,color = 'k',label = 'y = exp(x)')
plt.legend()
plt.show()
```
